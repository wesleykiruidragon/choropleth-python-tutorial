```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
from branca.element import Figure
from folium.plugins import MarkerCluster
```

```python
Crime=pd.read_csv("G:/PROJECT/Final_YearProjectüëå‚ù§/DATA/Crime_hotspot_Nairobi.csv")
```

```python
Crime.head()
```

```python
Crime.plot.scatter(x='Longitude', y='Latitude')
```

```python
figsize=Figure(height=400,width=900)
Web_map=folium.Map(location=[-1.32750,36.7267],zoom_start=10,contol_scale=True)
figsize.add_child(Web_map)
#add full screen button to map
plugins.Fullscreen(position='topleft').add_to(Web_map)
#measure control
measure_control=plugins.MeasureControl(position='topleft',active_color='red',completed_color='red',
                                       primary_left_unit='miles')

draw=plugins.Draw(export=True)
Web_map.add_child(measure_control)
#define variambles
folium.TileLayer('Stamen Terrain').add_to(Web_map)
folium.TileLayer('Stamen Toner').add_to(Web_map)
folium.TileLayer('Stamen Water Color').add_to(Web_map)
folium.TileLayer('cartodbpositron').add_to(Web_map)
folium.TileLayer('cartodbdark_matter').add_to(Web_map)
folium.LayerControl().add_to(Web_map)
```

```python
for i, row in Crime.iterrows():
    Latitude=Crime. at[i, 'Latitude']
    Longitude=Crime. at[i, 'Longitude']
    Region=Crime. at[i, 'Region']
    Hotspot_Name=Crime. at[i, 'Hotspot_Name']
    Type_of_Crime=Crime. at[i, 'Type_of_Crime']
    Time=Crime. at[i, 'Time']
   
   
    popup='<b>''Region:'+(Crime.at[i,'Region'])+'<br>'+'<b>'+'Hotspot_Name:'+str (Crime.at[i,'Hotspot_Name'])+'<br>'+'Type_of_Crime:'+'<b>'+ str  (Crime.at[i,'Type_of_Crime']) +'<b>'+'<br>'+'Time:'+str(Crime.at[i,'Time'])
   
                
    folium.CircleMarker(location=[Latitude,Longitude],radius=3,color = 'red',popup=popup,zoom=2,
                        tooltip="<b><font color='red'>Alert!Alert!Alert!</font><br><font color='black'>There have been a lot of crimes in Nairobi<br> every day and police have come up<br> with hotspot areas in the City.<br>Click here to see those areas, type of<br>crime and time of the day they mostly occuring:</font>").add_to(Web_map)
    Web_map
```
